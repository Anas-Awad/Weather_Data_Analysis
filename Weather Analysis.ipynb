### Q. 16) Convert the datetime to date time type and then get the average temp for each month 

```python
df['Date/Time'] = pd.to_datetime(df['Date/Time'])
```

```python
month = {1:'January', 2:'February', 3: 'March', 4:'April', 5:'May', 6:'June', 7:'July',8:'August', 9:'September', 10:'October', 11:'November', 12:'December'}
```

```python
df['Month'] = df['Date/Time'].dt.month.apply(lambda x: month[x])
df
```

```python
# Answer
df.groupby('Month')['Temp_C'].mean().sort_values(ascending = False)
```

### determine the max temperature for each weather and see which has a cold weather and wich is hot

```python
x = df.groupby('Weather Condition')['Temp_C'].max().sort_values(ascending = False).head(10)
x
```

```python
hot = sns.barplot(data=df[df['Weather Condition'].isin(x.index)], x = 'Temp_C',y= 'Weather Condition',order=x.index,estimator=max )
hot.figure.suptitle('most 10 cold weahter')
```

```python
y = df.groupby('Weather Condition')['Temp_C'].min().sort_values(ascending = True).head(10)
y
```

```python
cold = sns.barplot(data=df[df['Weather Condition'].isin(y.index)], x = 'Temp_C',y= 'Weather Condition',order=y.index,estimator=min)
cold.figure.suptitle('most 10 cold weahter')
```
